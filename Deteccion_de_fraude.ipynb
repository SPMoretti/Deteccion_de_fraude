## Preprocesamiento

### Scaler

```python
#Escalo las columnas "Amount" y "Time"

from sklearn.preprocessing import RobustScaler 

rob_scaler = RobustScaler()

df['Amount_escalado'] = rob_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['Time_escalado'] = rob_scaler.fit_transform(df['Time'].values.reshape(-1,1))

df.drop(['Time','Amount'], axis=1, inplace=True)

amount_escalado = df['Amount_escalado']
time_escalado = df['Time_escalado']

df.drop(['Amount_escalado', 'Time_escalado'], axis=1, inplace=True)
df.insert(0, 'Amount_escalado', amount_escalado)
df.insert(1, 'Time_escalado', time_escalado)

df.head()
```

### PCA

```python
# Separo features y target

columns = df.columns.tolist()
columns = [c for c in columns if c not in ["Class"]]
target = "Class"
X = df[columns]
y = df[target]

print(X.shape)
print(y.shape)
```

```python
pca = PCA(0.95)  # 95% de la varianza explicada.
X_pca = pca.fit_transform(X) 

X_pca.shape
```

```python
varianza_cada_componente = pca.explained_variance_ratio_
print(varianza_cada_componente.sum())

varianza_explicada = pca.explained_variance_ratio_.cumsum()

plt.ylabel('% de Varianza total explicada')
plt.xlabel('Número usado de componentes')
plt.title('PCA Analysis')
plt.plot(varianza_explicada)
plt.xlim(-1,50)
plt.style.context('seaborn-whitegrid')
```

```python
df_pca=pd.DataFrame(X_pca)
df_pca
```

```python
df_final=pd.concat([df["Class"],df_pca], axis=1)
df_final=df_final.sample(frac=1, random_state=12)

df_final
```

Este dataset final contiene las 22 variables principales que explican el 95% de la varianza y el target "Class".

## Modelado

Usaremos un modelo simple como Decision Tree y uno complejo como RandomForest. Luego los compararemos.

```python
# Separo features y target
columns = df_final.columns.tolist()
columns = [c for c in columns if c not in ["Class"]]
target = "Class"
X = df_final[columns]
y = df_final[target]

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=12)

sm = SMOTE(random_state=2)
x_train_res, y_train_res = sm.fit_sample(X_train, y_train)
```

### Decision Tree

```python
dt= DecisionTreeClassifier(random_state=12)
```

```python
dt_scores = cross_val_score(dt ,x_train_res, y_train_res, cv=10, scoring="roc_auc")

print(dt_scores)
print("ROC_AUC promedio de validación:", dt_scores.mean())
```

```python
dt.fit(x_train_res, y_train_res)
```

```python
y_pred_dt=dt.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred_dt)
```

```python
print(classification_report(y_test, y_pred_dt))
```

```python
print("ROC_AUC de test:", roc_auc_score(y_test, y_pred_dt))
print("Accuracy de test:", accuracy_score(y_test, y_pred_dt))
```

### Random Forest

```python
rdf= RandomForestClassifier(random_state=12)
```

```python
rdf_scores = cross_val_score(rdf ,x_train_res, y_train_res, cv=10, scoring="roc_auc")

print(rdf_scores)
print("ROC_AUC de validación promedio:", rdf_scores.mean())
```

```python
rdf.fit(x_train_res, y_train_res)
```

```python
y_pred_rdf=rdf.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred_rdf)
```

```python
print(classification_report(y_test, y_pred_rdf))
```

```python
print("ROC_AUC de test:", roc_auc_score(y_test, y_pred_rdf))
print("Accuracy de test:", accuracy_score(y_test, y_pred_rdf))
```

## Conclusión

Como se ve, el modelo de Random Forest performa algo mejor en cuanto a ROC_AUC. Este quiere decir que genera menos Falsos negativos y Falsos positivos.

```python
modelos=["Decison Tree","Random Forest"]
ROC_AUC=[roc_auc_score(y_test, y_pred_dt),roc_auc_score(y_test, y_pred_rdf)]
                                                        
plt.title('COMPARACIÓN DE MODELOS')
plt.ylabel('ROC_AUC')

sns.barplot(modelos, ROC_AUC)
```
